### # 데이터프레임 시각화

```python
#데이터프레임 변수 생성
dict_data = {"철수":[1,2,3,4],"영희":[2,3,4,5],"민수":[3,4,5,6],"수진":[4,5,6,7]}
df = DataFrame(dict_data)
```

```python
df
```

### # 차트 그리기

```python
# DataFrame 에서 바로 Matplotlib 사용 가능 (편리)
```

```python
# 선그래프 
df.plot()
plt.show()
```

```python
# ↑ 관찰
#  x 축은 0, 1, 2, 3  인덱스가 설정되어 있다
#  y 축은 1 ~ 7  범위로 설정되어 있다
#  범례(legend) 가 컬럼
```

```python
# 막대그래프  .
df.plot.bar()
plt.show()
```

```python
# 가로막대그래프
df.plot.barh()
plt.show()
```

```python
# 히스토그램
df.plot.hist()
plt.show()
```

```python
# ↑ 관찰
#  막대의 중심이 일정치가 않다..
#  철수는 1 부터 정확히 시작했는데
#  영희는 2부터 시작이 아니라..
#  민주도  3의 왼쪽부터..

# 이렇게 나타나는 이유는 마지막에 7 이 들어갈 공간(계급구간)이 없던 것이다..
# 

# 이것은 어떻게 설정할까?  --> 구간설정을 해주어야 한다.
```

```python
# 히스토그램 구간설정

df.plot.hist(bins=range(1,9,1))  # 1 ~ 8 까지의 숫자가 1step 씩 구간 형성
plt.show()
```

### # 차트에 옵션 추가하기

```python
# 기본 막대그래프
df.plot.bar()
plt.show()
```

```python
# 그래프 크기 설정
df.plot.bar(figsize=[10, 6])  # 10 : x  size, 6 : y size
plt.show()
```

```python
# 제목설정
df.plot.bar(figsize=[10, 6])
plt.title('예제')
plt.show()
```

```python
# 제목 폰트 크기 설정
df.plot.bar(figsize=[10, 6])
plt.title('예제', fontsize=18)  # fontsize=
plt.show()
```

```python
# x축 이름 설정
df.plot.bar(figsize=[10, 6])
plt.title('예제', fontsize=18)
plt.xlabel('xlabel')
plt.show()
```

```python
# x축 이름 및 폰트크기 설정
df.plot.bar(figsize=[10, 6])
plt.title('예제', fontsize=18)
plt.xlabel('xlabel', fontsize=16)
plt.show()
```

```python
# y축 이름 및 폰트 크기 설정 
df.plot.bar(figsize=[10, 6])
plt.title('예제', fontsize=18)
plt.xlabel('xlabel', fontsize=16)
plt.ylabel('ylabel', fontsize=16)
plt.show()
```

```python
# x축 눈금설정
# 설정할 눈금의 위치, 눈금의 이름, 폰트사이즈, 각도
df.plot.bar(figsize=[10, 6])
plt.title('예제', fontsize=18)
plt.xlabel('xlabel', fontsize=16)
plt.ylabel('ylabel', fontsize=16)
plt.xticks([0, 1, 2], ['첫째','둘째','셋째'], fontsize=10, rotation=0)   # x축 3개의 눈금만 설정
plt.show()
```

```python
# y축 눈금설정
df.plot.bar(figsize=[10, 6])
plt.title('예제', fontsize=18)
plt.xlabel('xlabel', fontsize=16)
plt.ylabel('ylabel', fontsize=16)
plt.xticks([0, 1, 2], ['첫째','둘째','셋째'], fontsize=10, rotation=0)   # x축 3개의 눈금만 설정
plt.yticks([1, 3, 5, 7], ['첫째', '셋째', '다섯째', '일곱번째'])
plt.show()
```

```python
# x축 범위설정
df.plot.bar(figsize=[10, 6])
plt.title('예제', fontsize=18)
plt.xlabel('xlabel', fontsize=16)
plt.ylabel('ylabel', fontsize=16)
plt.xticks([0, 1, 2], ['첫째','둘째','셋째'], fontsize=10, rotation=0)
plt.yticks([1, 3, 5, 7], ['첫째', '셋째', '다섯째', '일곱번째'])
plt.xlim([-1, 4])  # -1 ~ 4
plt.show()
```

```python
# y축 범위설정
df.plot.bar(figsize=[10, 6])
plt.title('예제', fontsize=18)
plt.xlabel('xlabel', fontsize=16)
plt.ylabel('ylabel', fontsize=16)
plt.xticks([0, 1, 2], ['첫째','둘째','셋째'], fontsize=10, rotation=0)
plt.yticks([1, 3, 5, 7], ['첫째', '셋째', '다섯째', '일곱번째'])
plt.xlim([-1, 4])  
plt.ylim([-1, 8])  # -1 ~ 8 범위
plt.show()
```
