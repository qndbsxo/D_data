````python
response = requests.get(urls[0])    
````

````python
# ↑ 위의 경우는 '정보가 전달되는 시점' 이 '즉시 전달'이 된다.

# 나중에 다양한 형태로 전달받은 데이터를 볼수 있다
# response.text, response.content, response.raw...  등..

# 위와 같은 데이터를 실질적으로 부르는 시점에
# 정보를 가져오도록 하는 옵션 stream !
````

````python
response = requests.get(urls[0], stream=True)
# response.text, response.content, response.raw ...    <-- response 의  body 데이터, 단, 어떠한 형태로 가져올지의 차이

# 그래서 일반적으로 '파일 다운로드' 시에는 위와 같이 stream=True 옵션을 넣어주자   (라고 생각해주자)
````

````python
# 다운받은 데이터 확인 (바이너리 형태디!)
response.content[:100]
````

````python
fp = open("./out/image.jpg", "wb")  # w => write, b => binary     b가 없으면  text 로 간주된다.
fp.write(response.content)
fp.close()
````

### 실습:
urls 안에 있는 모든 url 의 데이터를 images 폴더 안에 저장하기<br>
#### 08_21 12:00

````python
import os

# os.path.isdir("images")
# os.path.exists("images")

# images 폴더가 없으면 생성하기
if not os.path.exists("images"):
    os.makedirs("images")   
    print("images 폴더 생성")
````

````python
# 저장할 파일명도 변경이 되어야 겠죠?
# 가져올 url 에서 유추할수 있으면 좋겠죠.

# map 을 사용하는게 이 경우에는 괜찮다.    각 데이터별로 똑같은 함수 적용
# map : 함수(url => ________), 데이터(urls)
````

````python


def save_image(url):
    filename = url.split("/")[-1]  # '/' 로 나눈뒤 맨~뒤에 있는 것이 파일 이름
    response = requests.get(url, stream=True)
    fp = open("./images/{filename}".format(filename=filename), "wb")
    fp.write(response.content)
    fp.close() 
    
    return filename

# "http://m.blog.naver.com/blogyourlife/thumb.jpg".split("/")
````

````python
list(map(save_image, urls))
````
